```python
# If the 'released' column contains dates like "Month Day, Year", keep only the first part before the comma
movie_df['released']=movie_df['released'].str.split(',').str[0]
```

```python
# Reverse the order of words in the 'released' column (e.g., "January 1" becomes "1 January")
movie_df['released'] = movie_df['released'].apply(lambda x: ' '.join(str(x).split()[::-1]) if pd.notnull(x) else x)
```

```python
# Extract the day as 'released_date' (first word) and the month as 'released_month' (second word)

movie_df['released_date']=movie_df['released'].str.split(' ').str[0]
movie_df['released_month']=movie_df['released'].str.split(' ').str[1]
```

```python
movie_df.rename(columns={'year': 'released_year'}, inplace=True)
```

```python
# Remove spaces from director and star names to make them single words (e.g., "Steven Spielberg" -> "StevenSpielberg")
movie_df['director'] = movie_df['director'].str.replace(' ', '', regex=False)
movie_df['star'] = movie_df['star'].str.replace(' ', '', regex=False)
```

```python
movie_df.drop(columns=['rating','score','runtime','released','released_date'],inplace=True)
```

```python
# Import vectorizers for converting text data into numerical vectors

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer

# Create a CountVectorizer object with a maximum of 5000 features and English stop words removed
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python

# Transform the 'tags' text into a numerical array (bag-of-words model)
vector = cv.fit_transform(movie_df['tags']).toarray()
```

```python
vector.shape
```

```python
cv.get_feature_names_out()
```

```python
import nltk
```

```python
# Import NLTK's PorterStemmer for stemming words (reducing words to their root form)
from nltk.stem import PorterStemmer
Ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(Ps.stem(i))
    return " ".join(y)
```

```python
stem('drama stanleykubrick jacknicholson june')
```

```python
movie_df['tags']=movie_df['tags'].apply(stem)
```

```python
# Import cosine similarity function to measure similarity between movies
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
similarity.shape
```

```python
similarity[1]
```

```python
similarity[1].shape
```

```python
# Define a function to recommend similar movies based on cosine similarity
def recommend(movie):
        # Find the index of the movie in the DataFrame

    index = movie_df[movie_df['name'] == movie].index[0]
    
     # Get similarity scores for this movie, sorted in descending order
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    
     # Print the names of the top 5 most similar movies (excluding the movie itself)
    for i in distances[1:6]:
        print(movie_df['name'].iloc[i[0]])
        # print(i)  # Uncomment to see index and similarity score
```

```python
recommend('The Shining')
```

```python
recommend('The Best of Youth')
```

```python
recommend('The Dark Knight')
```
